```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/gold-price/deccan gold mines ltd eod price(da_xu_ly).csv')
df.Date = pd.to_datetime(df.Date,dayfirst=True)
df.set_index('Date',inplace=True)
```

```python
#Correlation Map 

plt.figure(figsize = [15, 7], clear = True, facecolor = '#EAEAE6')
sns.heatmap(df.corr(), annot = True, square = False, linewidths = 5,
            linecolor = "white", cmap = "Oranges");
```


Nhận Xét:
 
* Từ đồ thị có thể thấy thuộc tính Open và Close có mối tương quan thuận với nhau
* Ngược lại, Deliverable Quantity và % Deli. Qty to Traded Qty có mối tương quan khá ít

## Dưới đây là thể hiện rõ ràng sự tương quan sau khi sử dụng heatmap để hình dung sơ bộ sự tương quan của dữ liệu

```python
sns.jointplot(x = "Open", y = "Close", data = df, height = 8, ratio = 6, kind = "scatter");
```

```python
sns.jointplot(x = "Deliverable Quantity", y = "% Deli. Qty to Traded Qty", data = df, height = 8, ratio = 6, kind = "scatter");
```

```python
## targeted period
data = df[df.index >= '2019']
```

```python
fig,axes = plt.subplots(2,2,figsize=[15,7])

## resampling to daily freq (original data)
axes[0,0].plot(data.Open)
axes[0,0].set_title("Daily",size=16)

## resampling to monthly freq 
axes[0,1].plot(data.Open.resample('M').mean())
axes[0,1].set_title("Monthly",size=16)

## resmapling to quarterly freq 
axes[1,0].plot(data.Open.resample('Q').mean())
axes[1,0].set_title('Quarterly',size=16)

## resampling to annualy freq
axes[1,1].plot(data.Open.resample('A').mean())
axes[1,1].set_title('Annualy',size=16)

plt.tight_layout()
plt.show()
```

Nhận Xét:

* Giai đoạn trước năm 2020, Open price có dấu hiệu sụt giảm tới năm 2021 là chạm đáy.
* Sau đó thì Open price bắt đầu tăng trưởng ổn định trở lại
* Các kỹ thuật được áp dụng: Facet (Thể hiện biểu đồ qua từng ngày, từng tháng, từng quý, từng năm). Khi ta áp dụng kỹ thuật này sẽ cho ta dễ thấy được sự biến động cụ thể Open price của vàng qua từng khung thời gian mà ta thể hiện.
* Từ việc trực quan hóa, biểu đồ trên giúp ta thấy được Open price đang phục hồi khá nhiều sau Covid-19 từ đó giúp ích cho việc đầu tư.
