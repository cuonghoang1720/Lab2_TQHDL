```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/gold-price/deccan gold mines ltd eod price(da_xu_ly).csv')
df.Date = pd.to_datetime(df.Date,dayfirst=True)
df.set_index('Date',inplace=True)
```

```python
#Correlation Map 

plt.figure(figsize = [15, 7], clear = True, facecolor = '#EAEAE6')
sns.heatmap(df.corr(), annot = True, square = False, linewidths = 5,
            linecolor = "white", cmap = "Oranges");
```


Nhn Xt:
 
* T  th c th thy thuc tnh Open v Close c mi tng quan thun vi nhau
* Ngc li, Deliverable Quantity v % Deli. Qty to Traded Qty c mi tng quan kh t

## Di y l th hin r rng s tng quan sau khi s dng heatmap  hnh dung s b s tng quan ca d liu

```python
sns.jointplot(x = "Open", y = "Close", data = df, height = 8, ratio = 6, kind = "scatter");
```

```python
sns.jointplot(x = "Deliverable Quantity", y = "% Deli. Qty to Traded Qty", data = df, height = 8, ratio = 6, kind = "scatter");
```

```python
## targeted period
data = df[df.index >= '2019']
```

```python
fig,axes = plt.subplots(2,2,figsize=[15,7])

## resampling to daily freq (original data)
axes[0,0].plot(data.Open)
axes[0,0].set_title("Daily",size=16)

## resampling to monthly freq 
axes[0,1].plot(data.Open.resample('M').mean())
axes[0,1].set_title("Monthly",size=16)

## resmapling to quarterly freq 
axes[1,0].plot(data.Open.resample('Q').mean())
axes[1,0].set_title('Quarterly',size=16)

## resampling to annualy freq
axes[1,1].plot(data.Open.resample('A').mean())
axes[1,1].set_title('Annualy',size=16)

plt.tight_layout()
plt.show()
```

Nhn Xt:

* Giai on trc nm 2020, Open price c du hiu st gim ti nm 2021 l chm y.
* Sau  th Open price bt u tng trng n nh tr li
* Cc k thut c p dng: Facet (Th hin biu  qua tng ngy, tng thng, tng qu, tng nm). Khi ta p dng k thut ny s cho ta d thy c s bin ng c th Open price ca vng qua tng khung thi gian m ta th hin.
* T vic trc quan ha, biu  trn gip ta thy c Open price ang phc hi kh nhiu sau Covid-19 t  gip ch cho vic u t.
